```python
#finding spam emails from a dataset usiing pandas and scikit-learn

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
print("total samples", df.shape) #giving the total number of samples in the dataset
df.Category.value_counts()
```

```python
'''df['spam'] = df.Category.map({'ham': 0, 'spam': 1}) #creating a new column 'spam' with 0 for ham and 1 for spam
df.spam.value_counts()
                                         OR
'''

df['spam'] = df['Category'].apply(lambda x: 1 if x== 'spam' else 0) #creating a new column 'spam' with 0 for ham and 1 for spam
df.spam.value_counts()
```

```python
df.head()  #in this we have another column 'spam' which has 0 for ham and 1 for spam
```

```python
#training and testing the model
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['Message'], df['spam'], test_size= 0.2)
```

```python
type(X_train), type(y_train), type(X_test), type(y_test)  #checking the types of the variables
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
  #checking the shapes of the training and testing sets
```

```python
from sklearn.feature_extraction.text import CountVectorizer #converting text to vectors using CountVectorizer

cv = CountVectorizer()

X_train_cv = cv.fit_transform(X_train.values)  #fitting and transforming the training data
type(X_train_cv), type(X_train.values)  #checking the type 
X_train_cv
```

```python
#to view the vocabulary created by CountVectorizer, we have to convert it to a array first 
X_train_cv.toarray()[:][1]  #viewing the first 5 rows of the transformed training data
```

```python
# doing some more testing with CountVectorizer
#dir(cv)[20:50]  #checking the methods and attributes of CountVectorizer
cv.get_feature_names_out()[4444]  #viewing the first 10 features created by CountVectorizer
#cv.vocabulary_  #viewing the vocabulary created by CountVectorizer

```

```python
X_train_np = X_train_cv.toarray()  #converting the sparse matrix to a dense matrix

X_train_np[23][444]  #viewing the first row of sparse matrix 

```

```python
# to check where exactly the non zero values are in the sparse matrix using numpy

np.where(X_train_np[1] != 0)
```

```python
X_test_cv = cv.transform(X_test.values) #transforming the X_test data into the vectors
```

```python
from sklearn.naive_bayes import MultinomialNB 

model = MultinomialNB()  #creating a model using Multinomial Naive Bayes
```

```python
model.fit(X_train_cv, y_train) #fitting the model with training data . We are using X_train_cv because it is in a vectorised form  
```

```python
 
y_pred = model.predict(X_test_cv)
 
```

```python
from sklearn.metrics import classification_report  #this is for the better performance of the model 
```

```python
print(classification_report(y_test, y_pred) ) #first the test data then the predicted data )

```

```python
emails_cv = cv.transform(emails) #transforming the sample emails to vectorised form using the same vectorizer 'cv'
email_pred = model.predict(emails_cv) #predicting the sample emails using the trained model
```
